```python
pca = PCA(n_components=2)
pca.fit(np.concatenate([mu,zs],axis=0))
pca_mu = pca.transform(mu)
pca_zs = pca.transform(zs)
```

```python
plt.scatter(pca_mu[:,0], pca_mu[:,1],c = 'b',label='Old')
plt.legend()
plt.show()
```

```python
plt.scatter(pca_zs[:,0], pca_zs[:,1],c = 'r',label='New')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
ax = plt.axes()
plt.scatter(pca_mu[:,0], pca_mu[:,1],c = 'b',label='Old')
plt.scatter(pca_zs[:,0], pca_zs[:,1],c = 'r',label='New')
ax.set_xlim(-15,-10)
ax.set_ylim(-5,2)
plt.legend()
plt.show()
```
